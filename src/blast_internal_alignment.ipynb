```python
import networkx as nx 
import pandas as pd 
import plotly.graph_objects as go
```

```python
data = pd.read_csv('../output/test_alignment.csv')
# data = data[data['pident'] < 100]
Graph = nx.Graph()

queries = data['qseqid'].to_list()
targets = data['sseqid'].to_list()
perc_id = data['pident'].to_list() 

for i in range(len(perc_id)):
    Graph.add_node(queries[i], label=queries[i])
    Graph.add_node(targets[i], label=targets[i])
    Graph.add_edge(queries[i], targets[i])
    Graph[queries[i]][targets[i]]['weight'] = perc_id[i]

pos = nx.spring_layout(Graph)
    # Assign the pos attribute to the nodes
nx.set_node_attributes(Graph, pos, 'pos')
```

```python
edge_x = []
edge_y = []
for edge in Graph.edges():
    x0, y0 = Graph.nodes[edge[0]]['pos']
    x1, y1 = Graph.nodes[edge[1]]['pos']
    edge_x.append(x0)
    edge_x.append(x1)
    edge_x.append(None)
    edge_y.append(y0)
    edge_y.append(y1)
    edge_y.append(None)
    

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=2, color='#888'),
    hoverinfo='none',
    mode='lines')

node_x = []
node_y = []
for node in Graph.nodes():
    x, y = Graph.nodes[node]['pos']
    node_x.append(x)
    node_y.append(y)

node_trace = go.Scatter(
    x=node_x, y=node_y,
    mode='markers',
    hoverinfo='text',
    marker=dict(
        showscale=True,
        # colorscale options
        #'Greys' | 'YlGnBu' | 'Greens' | 'YlOrRd' | 'Bluered' | 'RdBu' |
        #'Reds' | 'Blues' | 'Picnic' | 'Rainbow' | 'Portland' | 'Jet' |
        #'Hot' | 'Blackbody' | 'Earth' | 'Electric' | 'Viridis' |
        colorscale='YlGnBu',
        reversescale=True,
        color=[],
        size=10,
        colorbar=dict(
            thickness=10,
            title='Node Connections',
            xanchor='left',
            titleside='right'
        ),
        line_width=2))
node_adjacencies = []
node_text = []
# for node, adjacencies in enumerate(Graph.adjacency()):
#     node_adjacencies.append(len(adjacencies[1]))
    # node_text.append('# of connections: '+str(len(adjacencies[1])))
sp = list(set([id.split('|')[0] for id in data['qseqid'].to_list()]))
for node in Graph.nodes:
    node_text.append(Graph.nodes[node]['label'].split('|')[0]+' | '+Graph.nodes[node]['label'].split('|')[2]+' | '+Graph.nodes[node]['label'].split('|')[-1])
    node_adjacencies.append(sp.index(Graph.nodes[node]['label'].split('|')[0]))
node_trace.marker.color = node_adjacencies
node_trace.text = node_text

fig = go.Figure(data=[edge_trace, node_trace],
             layout=go.Layout(
               #  title='<br>Network graph made with Python',
                title='BLASTn percent identicality between gene sequences',
                titlefont_size=16,
                showlegend=False,
                hovermode='closest',
                margin=dict(b=20,l=5,r=5,t=40),
                template='simple_white',
                xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))
                )
fig.show()
fig.write_html('../output/BLASTn_nx_no-filter_pident_shell.html')
```
